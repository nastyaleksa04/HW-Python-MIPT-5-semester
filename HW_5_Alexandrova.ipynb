Тогда сам список содержит в себе лишь ссылку на первую вершину, и вспомогательные данные вроде текущей длины). Список должен поддерживать обращение по индексам, красиво печататься, выдавать длину через len() и иметь работающие методы .pop(), .append(). Также список должен поддерживать итерацию по нему. *Можете также добавить .insert() и .pop() по индексу.

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self._next = None

    @property
    def next(self):
        return self._next
    @next.setter
    def next(self, next):
        if not isinstance(next, Node) and next != None:
            raise TypeError
        else:
            self._next = next
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        self._length = 0

    def __getitem__(self, index):
        node = self.head
        if index >= self._length or index < 0:
            raise IndexError('list index out of range')
        
        for i in range(index):
            node = node.next 
            
        return node.value
    
    def __str__(self):
        if (self._length == 0):
            return ''
        
        node = self.head
        string = ''
        for i in range(self._length - 1):
            string += str(node.value)
            string += str(' --> ')
            node = node.next
        string += str(node.value)

        return string
    
    def __len__(self):
        return self._length
    
    def pop(self, index=None):
        if (index is None):
            index = self._length - 1
        if index >= self._length or index < 0:
            raise IndexError('pop index out of range')

        if (index == 0):
            node = self.head
            node_value = node.value
            self.head = node.next
            del node
            self._length -= 1
            return node_value
        
        node = self.head
        for i in range(index - 1):
            node = node.next
        node_target = node.next
        node.next = node.next.next
        node_value = node_target.value
        del node_target
        self._length -= 1
        return node_value
    
    def append(self, other):
        if self.head == None:
            self.head = Node(other)
            self._length += 1
            
        else:
            node = self.head
            for i in range(self._length - 1):
                node = node.next
            node.next = Node(other)
            self._length += 1

    def insert(self, other, index=None):
        if (index is None):
            index = self._length
        if index > self._length or index < 0:
            raise IndexError('index out of range')

        if (index == 0):
            full_list = self.head
            self.head = Node(other)
            self.head.next = full_list
            self._length += 1
            
        else:
            node = self.head
            for i in range(index - 1):
                node = node.next
            node_target = Node(other)
            full_list = node.next
            node.next = node_target
            node.next.next = full_list
            
            self._length += 1


    def __iter__(self):
        self.flag = self.head
        return self

    def __next__(self):
        if (self.flag == None):
            raise StopIteration
        value = self.flag.value
        self.flag = self.flag.next
        return value
```

Реализовать структуру наследования классов геометрических фигур shape. Каждый класс должен обладать методами .area()_ и .perimeter() для вычисления площади и периметра соответственно, а также метод __str__(). Среди обязательных для реализации структур: круг, треугольник, прямоугольник, квадрат, ромб. Для простоты часть фигур можно конструировать из точек, передающихся в порядке обхода фигуры по часовой стрелке.

```python
class Shape:
    def __init__(self, name):
        self._area = None
        self._perimeter = None
        self.name = name
    
    @property
    def area(self):
        return self._area
    @property
    def perimeter(self):
        return self._perimeter
    
    def __str__(self):
        return 'Type of figure is:' + self.name +'. '+'Area of figure:'+ str(self.area) +'. '+'Perimeter of figure:' + str(self.perimeter)
    
```

```python
class Circle(Shape):
    def __init__(self, radius=1, x=0, y=0, name='circle'):
        super().__init__(name)
        self.r = radius
        self.coords = (x, y)
        self.update_parameters()

    @property
    def r(self):
        return self._r
    @r.setter
    def r(self, radius):
        if radius < 0 or (type(radius) != int and type(radius) != float):
            raise ValueError
        else:
            self._r = radius
        self.update_parameters()

    @property
    def coords(self):
        return self._coords
    @coords.setter
    def coords(self, coords):
        if type(coords) != tuple or ((type(coords[0]) != int and type(coords[0]) != float) or (type(coords[1]) != int and type(coords[1]) != float)):
            raise ValueError
        else:
            self._coords = coords
        
    def update_parameters(self):
        self._area =  3.14159*self._r**2
        self._perimeter = 2*3.14159*self._r
```
